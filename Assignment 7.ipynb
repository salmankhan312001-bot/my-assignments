Q2. Create a Python function called currency_converter() that takes two arguments:

- amount: The amount of money to be converted (in USD).
- exchange_rate (default is 280): The conversion rate from USD to another currency (e.g., PKR). You can assume the default rate is 1 USD = 280 PKR.
- 
The function should return a message showing how much the given amount is worth in the target currency (PKR).

Expected Output:

'100 USD is equivalent to 28000 PKR'

```python
# Function to convert USD to PKR
def currency_converter(amount, exchange_rate=280):
    converted_amount = amount * exchange_rate
    return f"{amount} USD is equivalent to {converted_amount} PKR"

# Test the function
print(currency_converter(100))
```

Q5. Write a Python function calculate_rectangle that accepts the length and width of a rectangle as arguments. Inside this function, create two nested functions:

- calculate_area: This function should calculate the area of the rectangle.
- calculate_perimeter: This function should calculate the perimeter of the rectangle.
- The main function should return both the area and the perimeter of the rectangle.

Expected Output:

Area: 15

Perimeter: 16


```python
# Function to calculate the area and perimeter of a rectangle
def calculate_rectangle(length, width):

    # Nested function to calculate area
    def calculate_area():
        return length * width

    # Nested function to calculate perimeter
    def calculate_perimeter():
        return 2 * (length + width)

    # Return both area and perimeter
    return calculate_area(), calculate_perimeter()


# Test the function
area, perimeter = calculate_rectangle(10, 5)

print("Area:", area)
print("Perimeter:", perimeter)
```

Q6. Create a list of numbers, generate a new list containing only the numbers that are divisible by 3 or 5. (List Comprehension)

```python
# Original list of numbers
numbers = [10, 15, 18, 20, 21, 25, 30, 33, 40, 45]

# List comprehension to get numbers divisible by 3 or 5
divisible_numbers = [num for num in numbers if num % 3 == 0 or num % 5 == 0]

# Display the result
print("Numbers divisible by 3 or 5:", divisible_numbers)
```

Q7:
Write a Python program that:

Has a balance of 1000.

Asks the user for a withdrawal amount.

Checks for these errors:

Non-numeric input → print "Please enter a valid number."

Amount ≤ 0 → print "Amount must be more than zero."

Amount > balance → print "Insufficient balance."

If valid, deducts amount and shows remaining balance.

Always prints "Thank you for using the ATM."

Expected Output:

- Enter amount to withdraw: 1500

  Insufficient balance.

  Thank you for using the ATM.


- Enter amount to withdraw:  0

  Amount must be more than zero.

  Thank you for using the ATM.


- Enter amount to withdraw:  K


  Please enter a valid number.

  Thank you for using the ATM.

```python
# Initial balance
balance = 1000

try:
    # Ask the user for the withdrawal amount
    amount = float(input("Enter amount to withdraw: "))

    # Check if the amount is greater than zero
    if amount <= 0:
        print("Amount must be more than zero.")

    # Check if the amount exceeds the balance
    elif amount > balance:
        print("Insufficient balance.")

    # Deduct the amount and display the remaining balance
    else:
        balance -= amount
        print("Withdrawal successful.")
        print("Remaining balance:", balance)

# Handle non-numeric input
except ValueError:
    print("Please enter a valid number.")

# This always runs
finally:
    print("Thank you for using the ATM.")
```

Q8

Write a Python program that:

Has 5 seats available (available_seats = 5).

Asks the user how many tickets they want to book.

Checks for these cases:

Non-numeric input → print "Please enter a valid number."

Tickets ≤ 0 → print "You must book at least 1 ticket."

Tickets more than available seats → print "Not enough seats available."

If valid, confirms the booking by printing how many tickets were booked.

Always prints "Thank you for booking with us." at the end.

Expected Output:

- How many tickets do you want to book? 3

  3 ticket(s) booked successfully.

  Thank you for booking with us.

- How many tickets do you want to book? 0


  You must book at least 1 ticket.

  Thank you for booking with us.

- How many tickets do you want to book? 7

  Not enough seats available.

  Thank you for booking with us.

- How many tickets do you want to book? five

  Please enter a valid number.

  Thank you for booking with us.





```python
# Available seats
available_seats = 5

try:
    # Ask the user how many tickets they want to book
    tickets = int(input("How many tickets do you want to book? "))

    # Check if the number of tickets is valid
    if tickets <= 0:
        print("You must book at least 1 ticket.")

    # Check if enough seats are available
    elif tickets > available_seats:
        print("Not enough seats available.")

    # Confirm the booking
    else:
        print(f"{tickets} ticket(s) booked successfully.")

# Handle non-numeric input
except ValueError:
    print("Please enter a valid number.")

# This always runs
finally:
    print("Thank you for booking with us.")
```

Q9:

Define a class named Light with two methods:

turn_on() — prints: "The light is now ON."

check_status() — prints: "The light is bright and working."

Create an object of the Light class and call both methods.

Expected Output:

The light is now ON.

The light is bright and working.

```python
# Define the Light class
class Light:

    # Method to turn on the light
    def turn_on(self):
        print("The light is now ON.")

    # Method to check the light status
    def check_status(self):
        print("The light is bright and working.")


# Create an object of the Light class
light = Light()

# Call both methods
light.turn_on()
light.check_status()
```

Q10:

Define a class named Student that has the following:

Attributes: name, grade, and age.

A method introduce() that prints:

"My name is {name}, I am {age} years old."

A method show_grade() that prints:

"My grade is {grade}."

Create an object of the Student class with sample data (e.g., "Alice", "A", 16), and call both methods on the object.

Expected Output:

My name is Alice, I am 16 years old.

My grade is A.

```python
# Define the Student class
class Student:

    # Constructor to initialize attributes
    def __init__(self, name, grade, age):
        self.name = name
        self.grade = grade
        self.age = age

    # Method to introduce the student
    def introduce(self):
        print(f"My name is {self.name}, I am {self.age} years old.")

    # Method to show the grade
    def show_grade(self):
        print(f"My grade is {self.grade}.")


# Create an object of the Student class
student = Student("Alice", "A", 16)

# Call both methods
student.introduce()
student.show_grade()
```
