##### 5. text = 'hello python world'

##### Replace spaces with underscores in the string (e.g., "hello python world" → "hello_python_world").

##### Split the modified string (e.g., "hello_python_world" → ['hello', 'python', 'world']).

##### Join the list elements with a hyphen (-) between them (e.g., ['hello', 'python', 'world'] → "hello-python-world").

##### Print all three results

##### Hint: Use string methods

```python
text = "hello python world"
step1 = text.replace(" ","_")
print(step1)
step2 = step1.split("_")
print(step2)
step3 = "-".join(step2)
print(step3)
```

##### 7.  You have a product code that follows the format "SKU-12345-XYZ".

##### Find the first position of the dash (-).

##### Count the number of dashes (-) in the product code.

##### Remove the "SKU-" prefix and "-XYZ" suffix from the product code, Extract the numeric part 12345

##### Hint: Use String Methods

```python
code = "SKU-12345-XYZ"
dash = code.find("-")
print(dash)
count = code.count("-")
print(count)
numeric_part = code.removeprefix("SKU-").removesuffix("-XYZ")
print(numeric_part)
```

##### 9. text = "    Data Analysis with Python is very powerful    "

##### Perform the following tasks:

##### Remove extra spaces using strip().
##### Convert the string into title case using title().
##### Replace "Powerful" with "Important".
##### Find the position of the word "Python".

```python
text = " Data Analysis with Python is very powerful "
step1 = text.strip()
print(step1)
step2 = text.title()
print(step2)
step3 = text.replace("powerful","important")
print(step3)
step4 = text.find("Python")
print(step4)
```
