#### Question 1

##### Calculate and write the output of the following expression:
**30 + 4 × 3 − 6**

```python
#x = (30 + 4 × 3 − 6)

multiply = 4 * 3  
add = 30 + multiply
sub = add - 6

print(multiply)
print(add)
print(sub)
print()
print("final answer:",30 + 4 * 3 - 6)
```

#### Question 2 

##### Calculate and write the output of the following expression:
**100 ÷ 4 + 7 − 5**

```python
#y = 100 / 4 + 7 - 5 

div = 100 // 4 
add = div + 7 
sub = add - 5 

print(div)
print(add)
print(sub)
print()
print("final answer:",100//4+7-5)
```

#### Question 3

##### Create a variable containing the string value "100". Convert it into an integer and display its data type before and after conversion.

```python
value = "100"

# Before conversion
print("Before Conversion:")
print("Value:", value)
print("Data Type:", type(value))

# Convert to integer
value = int(value)
print()
# After conversion
print("After Conversion:")
print("Value:", value)
print("Data Type:", type(value))
```
